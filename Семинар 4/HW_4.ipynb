### 1.1. Скачать данные по ссылке https://www.kaggle.com/datasets/ionaskel/laptop-prices

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
df = pd.read_csv("kc_house_data.csv", encoding="windows-1251")
df
```

### 1.2 Изучите стоимости недвижимости

```python
plt.pyplot.hist(df["price"], bins=20)
plt.pyplot.title("Стоимости недвижимости")
plt.pyplot.xlabel("Стоимость недвижимости")
plt.pyplot.ylabel("Количество");
```

График говорит, что большинство домов имеют стоимость до 1 000 000 

```python
df_price_low = df[df["price"] <= 1000000]
plt.pyplot.hist(df_price_low["price"], bins=20)
plt.pyplot.title("Стоимости недвижимости")
plt.pyplot.xlabel("Стоимость недвижимости")
plt.pyplot.ylabel("Количество");
```

Убрав выбросы мы смогли определить, что средняя (медианная) стоимость недвижимости (без учёта выбросов) колеблется около 400 000. Проверим догадки:

```python
df["price"].median()
```

### 1.3 Изучите распределение квадратуры жилой

```python
plt.pyplot.hist(df["sqft_lot"], bins=20)
plt.pyplot.title("Квадратура жилья")
plt.pyplot.xlabel("Квадратура жилья")
plt.pyplot.ylabel("Количество");
```

По графику видно, что в большинстве случае площадь менее 100 000 

```python
df_sqft_low = df[df["sqft_lot"] <= 100000]
plt.pyplot.hist(df_sqft_low["sqft_lot"], bins=20)
plt.pyplot.title("Квадратура жилья")
plt.pyplot.xlabel("Квадратура жилья")
plt.pyplot.ylabel("Количество");
```

По второму графику всё ещё непонятно распределение площади, необходимо отрезать ещё выбросы

```python
df_sqft_lower = df[df["sqft_lot"] <= 20000]
plt.pyplot.hist(df_sqft_lower["sqft_lot"], bins=20)
plt.pyplot.title("Квадратура жилья")
plt.pyplot.xlabel("Квадратура жилья")
plt.pyplot.ylabel("Количество");
```

Убрав выбросы мы смогли определить, что средняя (медианная) площадь недвижимости (без учёта выбросов) колеблется около 8000. Проверим догадки:

```python
df["sqft_lot"].median()
```

### 1.4 Изучите распределение года постройки

```python
plt.pyplot.hist(df["yr_built"], bins=20)
plt.pyplot.title("Квадратура жилья")
plt.pyplot.xlabel("Квадратура жилья")
plt.pyplot.ylabel("Количество");
```

По данному графику можно увидеть, что большинство домов "свежие", наименьшее количество домов построено до 1940 года.<br>

## Задача 3

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

В первую очередь посмотрим корреляцию:

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm");
```

Высокая корреляция с параметрами: sqft_living, grade, sqft_above, sqft_living15. Оценим зависимости

```python
sns.jointplot(x=df["price"], y=df["sqft_living"], kind="reg");
```

Из предыдущего графика очевидно, есть прямая зависимость жилой площади от стоимости жилья. Чем больше жилая площадь, тем выше стоимость. Наблюдение очевидное, но теперь мы можем убедиться

Кажется, что площадь дома (общая) также должна влиять на стоимость. Корреляционный анализ зависимости не выявил. Проверим:

```python
sns.jointplot(x=df["price"], y=df["sqft_lot"], kind="reg");
```

Действительно, зависимость очень слабая. Попробуем с "помытыми" данными:

```python
sns.jointplot(x=df_price_low["price"], y=df_sqft_lower["sqft_lot"], kind="reg");
```

Из данного графика видно, что действительно никакой зависимости от общей площади нет. 

Проверим зависимость от grade: 

```python
grade_price = df.groupby("grade").agg({"price": "mean"})
plt.pyplot.plot(grade_price)
plt.pyplot.title("Стоимости недвижимости от состояния")
plt.pyplot.xlabel("Состояние")
plt.pyplot.ylabel("Стоимость");
```

Мы видим, что с увеличением состояния стоимость растет. Рост не совсем линейный, до оценки 9 рост можно назвать линейным, далее он становится в разы сильнее. Объяснить данное явление можно тем, что по предыдущему анализу мы определили, что большинство недвижимости в состоянии 6-9, следовательно 10-12 найти достаточно сложно, их мало. Отсюда и рост цены.

В заключение рассмотрим зависимость стоимости недвижимости от количество ванных / спален. Корреляция не показала сильной зависимости, проверим:
